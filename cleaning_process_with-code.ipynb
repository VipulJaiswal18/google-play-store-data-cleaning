```python
df = pd.read_csv('google_play_store_dataset.csv')
```

```python
df.drop_duplicates(inplace = True)
```

```python
df
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean(), inplace = True)
```

```python
df[df['Type'].isnull()].index
```

```python
df['Type'] = df['Type'].fillna(df['Type'].mode()[0] )
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Content Rating'] = df['Content Rating'].fillna(df['Content Rating'].mode()[0])
```

```python
df['Current Ver'] = df['Current Ver'].fillna('varies with device')
```

```python
df['Android Ver'] = df['Android Ver'].fillna('varies with device')
```

```python
df.sort_values(by = 'Installs' , ascending = False , inplace = True)
```
